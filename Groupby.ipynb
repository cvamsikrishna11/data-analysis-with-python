## The .agg() Method

```python
fortune = pd.read_csv('pandas/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby(['Sector'])
fortune.head(3)
```

```python
sectors.agg({'Revenue': ['sum', 'mean'],
            'Profits':'sum',
            'Employees':'mean'})
```

```python
sectors.agg(['size', 'sum', 'mean'])
```

## Iterating through Groups

```python
fortune = pd.read_csv('pandas/fortune1000.csv', index_col='Rank')
sectors = fortune.groupby(['Sector'])
fortune.head(3)
```

```python
sectors['Profits'].max()
```

```python
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue_company_in_grp = data.nlargest(1, columns='Revenue')
    df = df.append(highest_revenue_company_in_grp)
```

```python
df
```

```python
cities = fortune.groupby('Location')
```

```python
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
for city, data in cities:
    highest_revenue_company_in_city = data.nlargest(1, columns='Revenue')
    df = df.append(highest_revenue_company_in_city)
```
