## Filtering with .where() method
### if we want to keep the original df but need to show the required data then we can use where

```python
bond = pd.read_csv('pandas/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
mask = bond['Actor'] == 'Sean Connery'
bond[mask]
```

```python
# where will return complete DF, with non present values as NaN
bond.where(mask)
```

```python
bond.where(bond['Box Office'] > 800)
```

```python
mask2 = bond['Box Office'] > 800
```

```python
bond.where(mask & mask2)
```

## The .apply() method on columns

```python
bond = pd.read_csv('pandas/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + ' MILLIONS! '
```

```python
bond['Box Office'] = bond['Box Office'].apply(convert_to_string_and_add_millions)
```

```python
bond.head(5)
```

```python
bond['Bond Actor Salary'].apply(convert_to_string_and_add_millions)
```

```python
# perform apply on multiple columns
columns = ['Box Office', 'Bond Actor Salary', 'Budget']
for each_value in columns:
    bond[each_value] = bond[each_value].apply(convert_to_string_and_add_millions)
```

```python
bond.head()
```

## The .apply() method on the rows

```python
bond = pd.read_csv('pandas/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    if actor == 'Pierce Brosnan':
        return 'The Best'
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    else:
        return 'I have no clue'
```

```python
# operations will perform on row values, but axis should be columns
bond.apply(good_movie, axis='columns')
```
